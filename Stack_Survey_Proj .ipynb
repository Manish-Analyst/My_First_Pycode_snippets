```python
country_respondent
```

```python
country_use_Python=Country_grp['LanguageWorkedWith'].apply(lambda x:x.str.contains('Python').sum())
```

```python
country_use_Python
```

```python
python_df=pd.concat([country_respondent,country_use_Python],axis='columns',sort=False)
```

```python
python_df.columns=['Country_Resp','Python_Resp']
```

```python
python_df['%Python_Resp']=python_df['Python_Resp']/python_df['Country_Resp']*100
```

```python
python_df.reset_index(inplace=True)
```

```python
ForGraph=python_df.head(5)
```

```python
ForGraph
```

```python
plt.style.use('fivethirtyeight')
```

```python
plt.barh(ForGraph['index'],ForGraph['%Python_Resp'])
plt.title('Country VS % Python_Resp')
plt.xlabel("%Respond")
plt.tight_layout()
plt.show()
```
